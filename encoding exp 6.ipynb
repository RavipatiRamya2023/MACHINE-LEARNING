```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Load the dataset
file_path = "/iris(1).csv"
df = pd.read_csv(file_path)

# Display the original DataFrame
print("Original DataFrame:")
print(df.head())

# Applying Label Encoding to 'species' column
label_encoder = LabelEncoder()
df['species_label'] = label_encoder.fit_transform(df['species'])
print("\nDataFrame with Label Encoding:")
print(df.head())

# Applying One-Hot Encoding
onehot_encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = onehot_encoder.fit_transform(df[['species']])

# Convert the result to a DataFrame
one_hot_df = pd.DataFrame(one_hot_encoded, columns=onehot_encoder.get_feature_names_out(['species']))

# Concatenating original DataFrame with One-Hot Encoding results
final_df = pd.concat([df, one_hot_df], axis=1)
print("\nFinal DataFrame with One-Hot Encoding:")
print(final_df.head())
```
